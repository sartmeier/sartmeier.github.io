```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
beforeI = pd.DataFrame({"0.4m":0, "1m":0, "2m":0, "4.4m":0}, 
                       index=["d00","d01","d02","d03","d04","d05","d06","d07","d08","d09","d10","d11","d12",
                              "d13", "d14", "d15", "d16", "d17", "d18", "d19", "d20", "d21"])
beforeII = pd.DataFrame({"0.4m":0, "1m":0, "2m":0, "4.4m":0}, 
                       index=["d00","d01","d02","d03","d04","d05","d06","d07","d08","d09","d10","d11","d12",
                              "d13", "d14", "d15", "d16", "d17", "d18", "d19", "d20", "d21"])
beforeII
```

```python
afterI = pd.DataFrame({"0.4m":0, "1m":0, "2m":0, "4.4m":0}, 
                       index=["d00","d01","d02","d03","d04","d05","d06","d07","d08","d09","d10","d11","d12",
                              "d13", "d14", "d15", "d16", "d17", "d18", "d19", "d20", "d21"])
afterII = pd.DataFrame({"0.4m":0, "1m":0, "2m":0, "4.4m":0}, 
                       index=["d00","d01","d02","d03","d04","d05","d06","d07","d08","d09","d10","d11","d12",
                              "d13", "d14", "d15", "d16", "d17", "d18", "d19", "d20", "d21"])
afterII
```

```python
beforeI.iloc[0] = [0,0,0,0]
beforeI.iloc[0]
```

```python
def change_day(day, b1, b2, a1, a2):
    beforeI.iloc[day] = b1
    beforeII.iloc[day] = b2
    afterI.iloc[day] = a1
    afterII.iloc[day] = a2
```

```python
change_day(4, [0,1,0,0], [2,0,1,1], [0,0,0,0], [1,0,0,0])
```

```python
beforeI
```

```python
change_day(5, [1,1,0,1], [2,2,1,1], [0,1,0,0], [1,0,0,0])
change_day(6, [1,2,0,1], [2,2,1,1], [1,1,0,0], [1,0,0,0])
change_day(7, [1,2,1,1], [2,3,2,1], [2,1,0,1], [1,0,0,0])
change_day(8, [5,2,2,1], [2,3,3,1], [3,1,0,1], [2,0,0,0])
change_day(9, [13,6,2,1], [2,3,3,1], [3,1,0,1], [3,0,0,0])
```

```python
plt.plot(beforeI)
plt.xlabel("days since test")
plt.ylabel("number of colonies")
plt.xlim(0,9)
```

```python
beforeI.iloc[0]
```

```python
medians_before = beforeI
medians_before.iloc[:,0] = np.median(np.asarray(list(zip(beforeI.iloc[:,0],beforeII.iloc[:,0]))),axis=1)
medians_before.iloc[:,1] = np.median(np.asarray(list(zip(beforeI.iloc[:,1],beforeII.iloc[:,1]))),axis=1)
medians_before.iloc[:,2] = np.median(np.asarray(list(zip(beforeI.iloc[:,2],beforeII.iloc[:,2]))),axis=1)
medians_before.iloc[:,3] = np.median(np.asarray(list(zip(beforeI.iloc[:,3],beforeII.iloc[:,3]))),axis=1)
```

```python
medians_before
```

```python
medians_after = afterI
medians_after.iloc[:,0] = np.median(np.asarray(list(zip(afterI.iloc[:,0],afterII.iloc[:,0]))),axis=1)
medians_after.iloc[:,1] = np.median(np.asarray(list(zip(afterI.iloc[:,1],afterII.iloc[:,1]))),axis=1)
medians_after.iloc[:,2] = np.median(np.asarray(list(zip(afterI.iloc[:,2],afterII.iloc[:,2]))),axis=1)
medians_after.iloc[:,3] = np.median(np.asarray(list(zip(afterI.iloc[:,3],afterII.iloc[:,3]))),axis=1)
medians_after
```

```python
plt.figure(figsize=(16,5))
plt.plot(medians_before.iloc[:,0],linestyle="dashed",color='#377eb8',label="0,4m")
plt.plot(medians_after.iloc[:,0],color='#377eb8',label="0,4m")#'#ff7f00')
plt.plot(medians_before.iloc[:,1],linestyle="dashed",color='#4daf4a',label="1m")
plt.plot(medians_after.iloc[:,1],color='#4daf4a',label="1m")#'#f781bf')
plt.plot(medians_before.iloc[:,2],linestyle="dashed",color='#f781bf',label="2m")#'#a65628')
plt.plot(medians_after.iloc[:,2],color='#f781bf',label="2m")#'#a65628')#'#984ea3')
plt.plot(medians_before.iloc[:,3],linestyle="dashed",color='#ff7f00',label="4,4m")
plt.plot(medians_after.iloc[:,3],color='#ff7f00',label="4,4m")#'#999999')#'#e41a1c')
plt.legend()
plt.xlabel("days since test")
plt.ylabel("number of colonies")
plt.xlim(0,9)
```
